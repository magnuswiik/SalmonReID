# Find trajectories for each fish individual

```python
import os, json
import pandas as pd

boxes_path = "/Users/magnuswiik/prosjektoppgave_data/Masteroppgave_data/Helfisk_Deteksjonssett/Boxes/"
imgs_path = "/Users/magnuswiik/prosjektoppgave_data/Masteroppgave_data/Helfisk_Deteksjonssett/Images/"

from_img = "GP020101_00012069.jpg"
from_json = "GP020101_00012069.json"

annots = sorted(file for file in os.listdir(boxes_path) if file.endswith('.json'))
imgs = sorted(file for file in os.listdir(imgs_path) if file.endswith('.jpg'))

traj_df = pd.DataFrame(columns=['FishID', 'StartFrame', 'EndFrame', 'TrajLength', 'VideoID'])

fish_seen_last_time = {}

for i in range(len(imgs)):

    fish_in_frame = []

    with open(boxes_path + annots[i], 'r') as file:
        content = json.load(file)
        shapes = content['shapes']
        
        for shape in shapes:
            id = shape['group_id']
            video_id = annots[i].split("_")[0]
            traj_id = str(id) + "_" + video_id
            fish_in_frame.append(traj_id)
            if traj_id not in fish_seen_last_time.keys():
                fish_seen_last_time[traj_id] = {'startFrame': annots[i].split("_")[1][:-5],
                                                'seenCount': 0,
                                                'sinceSeen': 0}
        to_delete = []
        
        if annots[i].split("_")[1][:-5] == '00000331':
            print("")
        
        for fish, info in fish_seen_last_time.items():
            if fish not in fish_in_frame:
                fish_seen_last_time[fish]['sinceSeen'] += 1
                if fish_seen_last_time[fish]['sinceSeen'] >= 20:
                    endFrame = annots[i-5].split("_")[1][:-5]
                    traj_df.loc[len(traj_df)] = [fish, fish_seen_last_time[fish]['startFrame'], endFrame, fish_seen_last_time[fish]['seenCount'], annots[i].split("_")[0]]
                    to_delete.append(fish)
            else:
                fish_seen_last_time[fish]['sinceSeen'] = 0
                fish_seen_last_time[fish]['seenCount'] += 1
        
        for fish in to_delete:
            del fish_seen_last_time[fish]



traj_df.sort_values(by='FishID', ascending=True, inplace=True, ignore_index=True)
#traj_df.to_csv("fish_trajectories.csv", index=None)
traj_df                
```
